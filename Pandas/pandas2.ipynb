## Valores NaN

*Tambien conocidos como `Not a number` son valores no definidios que casi siempre son espacios en blanco , son muy vistos en este mundo pues son campos de informacion sin datos que en muchos casos tenemos que rellenar usando tecnicas, como nuestro dataset no tiene este tipo de valores, crearemos con estos datos para ver como rellenar la informacion*

```python
import pandas as np
import numpy as np

# Crear un DataFrame con NaNs
data = {
    'Nombre': ['Juan', 'María', 'Luis', np.nan, 'Juan', 'Lucía', 'Miguel', 'Sofía', 'María', np.nan],
    'Edad': [23, 21, np.nan, 22, 25, 20, 24, np.nan, 22, 21],
    'Calificación': [np.nan, 88, 92, np.nan, 85, 79, 90, np.nan, 77, 82],
    'Asistencia': [10, np.nan, 15, 14, np.nan, 13, np.nan, 12, 10, 11],
    'Proyecto': [np.nan, 'A', 'B', 'C', 'D', 'E', np.nan, 'G', 'H', np.nan]
}

df_nan = pd.DataFrame(data)


df_nan
```

Podemos evidenciar varias cosas,
- Hay espacios vacios de tipo NaN en el Nombre,Edad,Calificacion,Asistencia,Proyecto
- Podemos aplicar tecnicas para rellenar los valores usando estadistica
- En importante analizar cuando es bueno eliminar datos o rellenarlos.
- Existen tecnicas mejores y peores, todo es cuestion de ver cuando usarse

  
  

```python
#Aplicares la moda para los nombres

nombres_moda=df_nan['Nombre'].mode()

print(nombres_moda) 
# Me muestra la moda del a conjunto a analizar ,en donde esta en orden de decendente

print(df_nan['Nombre'].value_counts())

#Maria y juan son los nombres mas usados,como estan iguales en cantidad
    #podemos usar cualquier o el que nos quede a nuestra seleccion
    #es bueno analizar antes
```

```python
# Corrige los valores NaN en la columna 'Nombre' del DataFrame df_nan
df_nan['Nombre'] = df_nan['Nombre'].fillna(nombres_moda[0])

# Muestra el DataFrame actualizado
df_nan
```

*Algunas veces no es lo mas indicado hacer esto para los nombres pues se estan repitiendo datos y a la hora de alimentar un modelo ML puede haber un mayor peso cuando alguien posee estos nombres*

**Ahora vamos a llenar los valores de la edad usando la media de la edad que hay en esa columna

```python
edad_media=df_nan['Edad'].mean() #Calculamos la media de la edad

print(df_nan['Edad'].median()) #Tambien podemos usar la mediana , sin embargo en este caso dieron muy iguales

print(edad_media)

df_nan['Edad']=df_nan['Edad'].fillna(edad_media)

df_nan
```

```python
#Ahora terminemos de hallar las que faltan

calificacion_Media=df_nan['Calificación'].mean()

asistencia_Media=df_nan['Asistencia'].mean()

df_nan['Calificación']=df_nan['Calificación'].fillna(calificacion_Media)

df_nan['Asistencia']=df_nan['Asistencia'].fillna(asistencia_Media)

df_nan #Ya arreglemos la calificacion y la asistencia

```

*Ahora nos falta por ultimo la columna `Proyecto` que es necesario analizar que es , parece ser la asignacion de un grupo, lo mas indicado seria llenar los espacios con nuevos grupos*

```python
import pandas as pd


nuevosGrupos=['I','J','K']


iterador_grupos = iter(nuevosGrupos)

# Función para aplicar los valores de nuevosGrupos a NaN secuencialmente
def aplicar_grupo(val):
    if pd.isna(val):  # Verifica si el valor es NaN
        try:
            return next(iterador_grupos)  # Obtiene el siguiente valor del iterador
        except StopIteration:
            return val  # Si no hay más valores en el iterador, deja el valor como está (NaN o lo que sea)
    else:
        return val  # Si el valor no es NaN, lo deja como está

# Aplica la función a la columna 'Proyecto'
df_nan['Proyecto'] = df_nan['Proyecto'].apply(aplicar_grupo)

df_nan
```

*Con esto llenamos todos los espacios disponibles aunque tambien podemos hacerlo manual si lo queremos hacer*

```python
df_nan.iloc[0, df_nan.columns.get_loc('Proyecto')] = 'W' #Podemos rellenar un espacio con algo en especifico

df_nan
```

## One hot codding 

Esta es una tecnica del mundo de analisis de datos que nos permite facilitar la comprension y la alimentacion de un modelo de ML, se intenta convertir una variable en un 1 y 0, haciendo que los numeros a analizar sean muy diminutos y llevando una expresion de una columna a algo mas booleano

**Ejemplo**

| Fruta  | Manzana | Banana | Naranja |
|--------|---------|--------|---------|
| Manzana| 1       | 0      | 0       |
| Banana | 0       | 1      | 0       |
| Naranja| 0       | 0      | 1       |


La columna "Manzana" tiene un 1 donde la fruta es "Manzana" y 0 en las otras filas.

La columna "Banana" tiene un 1 donde la fruta es "Banana" y 0 en las otras filas.

La columna "Naranja" tiene un 1 donde la fruta es "Naranja" y 0 en las otras filas.

Cada columna representa una categoría específica y muestra si esa categoría está presente o no en cada fila. No hay dependencia de una columna en función de sí misma; más bien, cada columna es independiente y refleja la presencia o ausencia de una categoría particular para cada observación en el conjunto de datos.

-Por ejemplo vamos agregar una columna al dataset anterior basada en el genero anterior

```python
generos=["Hombre","Mujer","Hombre","Hombre","Hombre","Mujer","Hombre","Mujer","Mujer","Hombre"]

df_nan['Genero'] = generos

df_nan

df_nan=pd.get_dummies(df_nan,columns=['Genero']) #Con esto generamos una tabla con la respuesta para el genero

'''Pasa de un estado de string a un objetivo mas booleano que permite el procesamiento de datos mucho
mas facil haciendo que cuando se alimente sea mas facil el despliegue y desarollo del modelo'''

#Haremos un diminuto cambio para que cuando sea True ponga un 1 y false un 0

df_nan['Genero_Hombre'] = df_nan['Genero_Hombre'].astype(int) #Es posible convetir los booleanos a numeros automatico

df_nan['Genero_Mujer']=df_nan['Genero_Mujer'].astype(int)

df_nan
```
