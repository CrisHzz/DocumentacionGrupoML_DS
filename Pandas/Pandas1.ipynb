```python
import pandas as pd #importamos pandas
```

Vamos a leer uno de los archivos que ya ternemos , se trata de la lista de todos los equipos de futbol del mundo y tambien hay otro con los jugadores por si lo quieres curiosear

```python
file_path = r'../Media/male_players.csv' #La ruta del archivo ,podemos copiar y pegarla , se necesita que sea un string y se pone una r adelante para indicar su relatividad , es mas por conveccion

dataset=pd.read_csv(file_path)
```

Aqui estamos leyendo la ruta y usamos la funcion pd.read de pandas para leer el archivo , debemos de usar el indicado del archivo dependiendo del que sea , sea xlss , txt ,csv como es en este caso

Apartir de ahora llamaremos `dataset` a aquellos archivos que manejan datos y ya son leidos por pandas

```python
# Al ser tan grande el dataset, vamos a enfoncarnos en un equipo

nacional=dataset[dataset['club_name']=='Atlético Nacional']

nacional.head(9)
```

```python
nacional=nacional[['short_name','long_name','player_positions','overall','height_cm','weight_kg','value_eur','age','club_name']]

nacional.head(5)
```

```python
dim=dim[['short_name','long_name','player_positions','overall','height_cm','weight_kg','value_eur','age','club_name']]

dim.head(5)
```

```python
city=dataset[dataset['club_name']=='Manchester City']

city=city[['short_name','long_name','player_positions','overall','height_cm','weight_kg','value_eur','age','club_name']]

city.head(5)
```

```python
rma=dataset[dataset['club_name']=='Real Madrid']

rma=rma[['short_name','long_name','player_positions','overall','height_cm','weight_kg','value_eur','age','club_name']]

rma.head(10)
```
