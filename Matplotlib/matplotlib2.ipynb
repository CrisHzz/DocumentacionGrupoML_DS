# Tipos de graficos en matplotlib para analizar datos

*Ya conocimos el gr치fico b치sico de Matplotlib, una gr치fica simple que puede tener forma lineal, polinomial o la que sea dependiendo de lo que se quiera hacer, vamos a ver algunos tipos de gr치ficos para poder analizar datos*

## Grafico de barras

```python
import matplotlib.pyplot as plt

import numpy as np

perros = np.arange(5,51,5) # Cantidad de perros por comuna

gatos = [1, 3, 5, 12, 15, 16, 17, 18, 12, 22]  # Cantidad de gatos por comuna

# Crear un rango para el n칰mero de comunas (10)
comunas = np.arange(1, len(perros) + 1)

# Crear el objeto figura y el subplot

ancho_barra = 0.30

#Correr posicion de los elementos
pos_perros = comunas - ancho_barra/2
pos_gatos = comunas + ancho_barra/2

# Graficar las barras
plt.bar(pos_perros, perros, width=ancho_barra, label='Perros')
plt.bar(pos_gatos, gatos, width=ancho_barra, label='Gatos')

# Etiquetas y t칤tulo
plt.xlabel('Comunas')
plt.ylabel('Cantidad')
plt.title('Cantidad de Perros y Gatos por Comuna')
plt.legend()

# Mostrar la gr치fica
plt.show()
```

## Historigrama


```python

poblacionUdem = [100, 70, 150, 50, 23]  # Cantidad de chicas en la estaci칩n

dias = ["Lunes", "Martes", "Mi칠rcoles", "Jueves", "Viernes"]

# Crear el histograma
fig, ax = plt.subplots()

# Graficar el histograma
ax.hist(dias, weights=poblacionUdem, bins=len(dias), color="skyblue", alpha=0.5, label="Ocurrencia en la estaci칩n UDEM")
#Solamente recibe una variable y queda en y. Bins te mostrara la cantidad de contenedores de comparacion para los datos los dias de semana (5)
#Weights le da el peso a la comparacion de frecuencia en el histograma a evaluar
#Alpha le indica la opacidad
ax.set_xlabel("D칤as de la semana")
ax.set_ylabel("Cantidad de personas")
ax.set_title("Ocurrencia de personas en la estaci칩n UDEM")
ax.legend()

plt.show()
```

## Diagrama circular o de torta (Cake)

```python
import matplotlib.pyplot as plt

# Datos
stations = ["Udem", "Los alpes", "La palma", "Parque Belen", "Industriales"]
percentageStation = [20.5, 13.0, 8.3, 28.0, 30.2]
highlight = (0, 0, 0, 0, 0.1)
colors = ['salmon', 'aquamarine', 'mediumturquoise', 'deeppink', 'blueviolet']

# Crear el gr치fico de torta
plt.figure(figsize=(10, 8))
plt.title("Ocurrencia en estaciones por %")

wedges, texts, autotexts = plt.pie(
    percentageStation, explode=highlight, labels=stations, autopct="%1.1f%%",
    shadow=True, startangle=90, colors=colors
)

# A침adir la leyenda
plt.legend(wedges, stations, title="Estaciones", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Mostrar el gr치fico
plt.show()
```

## Diagrama de caja

Un diagrama de caja, tambi칠n conocido como diagrama de caja y bigotes o boxplot en ingl칠s, es una representaci칩n gr치fica que muestra la distribuci칩n de un conjunto de datos num칠ricos a trav칠s de sus cuartiles. 

El diagrama de caja se utiliza para visualizar la variabilidad, la simetr칤a y los valores at칤picos de un conjunto de datos. Proporciona informaci칩n sobre la mediana, los cuartiles, los valores m칤nimo y m치ximo, as칤 como los valores at칤picos potenciales.

El diagrama de caja consta de una caja rectangular que representa el rango intercuart칤lico (IQR), que abarca el 50% central de los datos.
