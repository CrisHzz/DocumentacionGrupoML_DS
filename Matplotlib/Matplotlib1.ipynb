*Ya sabemos leer datos y crear componentes para almacenarlos, ahora necesitamos visualizarlos, pues estos serían muy difíciles de ver si siempre vemos tablas y vectores*

*Aquí es donde llega `Matplotlib` será nuestra librería para la visualización de datos y le intentara darle imagen a todo lo que queremos mostrar, es fundamental para el análisis de datos y comportamiento de un modelo de ML*

```python
import matplotlib.pyplot as plt #Importacion de la libreria

import numpy as np

#Para poder visualizar datos primero los necesitamos vamos a crear algo basico

#Velocidad vs tiempo

tiempo=np.arange(1,11)

velocidad=np.arange(0,20,2)

plt.figure(figsize=(10,6)) #Este metodo me genera la figura de las dimensiones que quiero

#Puedo ver la documentacion para ver las dimensiones o a tanteo
```

Hay que entender primero que los datos tienen un sentido a la hora de gráfica, siempre en el eje de las `X` va las variables independientes, es decir, su obtención no depende de otro valores para llegar a su resultado y la dependiente va en el eje `Y`

```python
plt.plot(tiempo,velocidad) #Asi puedo mostrar el grafico de la manera mas simple posible

plt.show() #Esta funcion siempre va al final de un grafico para poder concretar su visualizacion
```

*Asi se crea de manera simple el grafico pero se ve muy feo , vamos a ponerlo bonito*

```python
plt.plot(tiempo,velocidad,color='purple',label='V/t') #color muestra el color,
#label mostrara la leyenda de los datos a analizar

plt.title('Grafico Velocidad vs Tiempo') #Titulo del grafico

plt.xlabel('Tiempo') #Eje x

plt.ylabel('velocidad') #Eje y

plt.legend() #Mostrar la leyenda

plt.show()
```

```python
plt.plot(tiempo,velocidad,color='purple',label='V/t',marker='D',linestyle='--')

#Podemos decorar las lineas y marcadores usando algunos parametros

plt.title('Grafico Velocidad vs Tiempo') #Titulo del grafico

plt.xlabel('Tiempo') #Eje x

plt.ylabel('velocidad') #Eje y

plt.legend() #Mostrar la leyenda

plt.grid(True) #Puedo verlo en forma de grindilla para visualizar de otra forma

plt.show()
```

## COLORES BASICOS DE MATPLOTLIB

```python
#TODOS LOS COLORES DE MATPLOTLIB

import math

import matplotlib.pyplot as plt

import matplotlib.colors as mcolors
from matplotlib.patches import Rectangle


def plot_colortable(colors, *, ncols=4, sort_colors=True):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        names = sorted(
            colors, key=lambda c: tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(c))))
    else:
        names = list(colors)

    n = len(names)
    nrows = math.ceil(n / ncols)

    width = cell_width * ncols + 2 * margin
    height = cell_height * nrows + 2 * margin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-margin)/height)
    ax.set_xlim(0, cell_width * ncols)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.add_patch(
            Rectangle(xy=(swatch_start_x, y-9), width=swatch_width,
                      height=18, facecolor=colors[name], edgecolor='0.7')
        )

    return fig

plot_colortable(mcolors.CSS4_COLORS)
plt.title('TODOS LOS COLORES BASICOS')
plt.show()
```

## LINEAS EN MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np

linestyle_str = [
     ('solid', 'solid'),      # Same as (0, ()) or '-'
     ('dotted', 'dotted'),    # Same as (0, (1, 1)) or ':'
     ('dashed', 'dashed'),    # Same as '--'
     ('dashdot', 'dashdot')]  # Same as '-.'

linestyle_tuple = [
     ('loosely dotted',        (0, (1, 10))),
     ('dotted',                (0, (1, 1))),
     ('densely dotted',        (0, (1, 1))),
     ('long dash with offset', (5, (10, 3))),
     ('loosely dashed',        (0, (5, 10))),
     ('dashed',                (0, (5, 5))),
     ('densely dashed',        (0, (5, 1))),

     ('loosely dashdotted',    (0, (3, 10, 1, 10))),
     ('dashdotted',            (0, (3, 5, 1, 5))),
     ('densely dashdotted',    (0, (3, 1, 1, 1))),

     ('dashdotdotted',         (0, (3, 5, 1, 5, 1, 5))),
     ('loosely dashdotdotted', (0, (3, 10, 1, 10, 1, 10))),
     ('densely dashdotdotted', (0, (3, 1, 1, 1, 1, 1)))]


def plot_linestyles(ax, linestyles, title):
    X, Y = np.linspace(0, 100, 10), np.zeros(10)
    yticklabels = []

    for i, (name, linestyle) in enumerate(linestyles):
        ax.plot(X, Y+i, linestyle=linestyle, linewidth=1.5, color='black')
        yticklabels.append(name)

    ax.set_title(title)
    ax.set(ylim=(-0.5, len(linestyles)-0.5),
           yticks=np.arange(len(linestyles)),
           yticklabels=yticklabels)
    ax.tick_params(left=False, bottom=False, labelbottom=False)
    ax.spines[:].set_visible(False)

    # For each line style, add a text annotation with a small offset from
    # the reference point (0 in Axes coords, y tick value in Data coords).
    for i, (name, linestyle) in enumerate(linestyles):
        ax.annotate(repr(linestyle),
                    xy=(0.0, i), xycoords=ax.get_yaxis_transform(),
                    xytext=(-6, -12), textcoords='offset points',
                    color="blue", fontsize=8, ha="right", family="monospace")


fig, (ax0, ax1) = plt.subplots(2, 1, figsize=(10, 8), height_ratios=[1, 3])

plot_linestyles(ax0, linestyle_str[::-1], title='Named linestyles')
plot_linestyles(ax1, linestyle_tuple[::-1], title='Parametrized linestyles')

plt.tight_layout()
```

El parámetro `figsize` en la función [`subplots`](command:_github.copilot.openSymbolFromReferences?%5B%7B%22%24mid%22%3A1%2C%22external%22%3A%22file%3A%2F%2F%2Fc%253A%2FUsers%2FCristian%2FDocuments%2FDocumentacionGrupoML_DS%2Fvenv%2FLib%2Fsite-packages%2Fmatplotlib%2Fpyplot.py%22%2C%22path%22%3A%22%2Fc%3A%2FUsers%2FCristian%2FDocuments%2FDocumentacionGrupoML_DS%2Fvenv%2FLib%2Fsite-packages%2Fmatplotlib%2Fpyplot.py%22%2C%22scheme%22%3A%22file%22%7D%2C%7B%22line%22%3A1553%2C%22character%22%3A4%7D%5D "venv/Lib/site-packages/matplotlib/pyplot.py") de Matplotlib se utiliza para especificar las dimensiones del objeto de figura (figure) que se va a crear. 

`figsize=(10, 6)` indica que se desea crear una figura con un ancho de 10 unidades y una altura de 5 unidades. Estas unidades pueden ser cualquier unidad de medida que desees utilizar, como pulgadas, centímetros, etc. 

```python
fig,grafica=plt.subplots(figsize=(10,6))

grafica.plot(tiempo,velocidad) # Ya en vez de usar plt usamos el nombre del objeto

grafica.set_xlabel('Tiempo')

grafica.set_ylabel('Velocidad')

grafica.set_title('Tiempo vs velocidad')

plt.show()
```

La función [`subplot`](command:_github.copilot.openSymbolFromReferences?%5B%7B%22%24mid%22%3A1%2C%22path%22%3A%22%2Fc%3A%2FUsers%2FCristian%2FDocuments%2FDocumentacionGrupoML_DS%2Fvenv%2FLib%2Fsite-packages%2Fmatplotlib%2Fpyplot.py%22%2C%22scheme%22%3A%22file%22%7D%2C%7B%22line%22%3A1365%2C%22character%22%3A4%7D%5D "venv/Lib/site-packages/matplotlib/pyplot.py") tiene varios parámetros que se pueden especificar entre paréntesis. Estos parámetros determinan la posición y las características del subgráfico que se va a crear en una figura.

Aquí hay una descripción de los diferentes valores que se pueden proporcionar entre paréntesis:

- `nrows`: El número de filas en la cuadrícula de subgráficos.
- `ncols`: El número de columnas en la cuadrícula de subgráficos.
- `index`: La posición del subgráfico en la cuadrícula, especificada como un número entero o una tupla de dos enteros. El índice comienza en 1 en la esquina superior izquierda y aumenta hacia la derecha.
- `pos`: Una forma abreviada de especificar la posición del subgráfico en la cuadrícula. Puede ser un número entero de tres dígitos, donde los dígitos se interpretan como números enteros separados, por ejemplo, `235` es equivalente a `2, 3, 5`.
- [`ax`](command:_github.copilot.openSymbolFromReferences?%5B%7B%22%24mid%22%3A1%2C%22fsPath%22%3A%22c%3A%5C%5CUsers%5C%5CCristian%5C%5CDocuments%5C%5CDocumentacionGrupoML_DS%5C%5CMatplotlib%5C%5CMatplotlib1.ipynb%22%2C%22_sep%22%3A1%2C%22path%22%3A%22%2Fc%3A%2FUsers%2FCristian%2FDocuments%2FDocumentacionGrupoML_DS%2FMatplotlib%2FMatplotlib1.ipynb%22%2C%22scheme%22%3A%22vscode-notebook-cell%22%2C%22fragment%22%3A%22X22sZmlsZQ%3D%3D%22%7D%2C%7B%22line%22%3A0%2C%22character%22%3A5%7D%5D "c:\Users\Cristian\Documents\DocumentacionGrupoML_DS\Matplotlib\Matplotlib1.ipynb"): Un objeto `Axes` existente que se desea utilizar en lugar de crear uno nuevo.

## Comparacion de graficas

*Podemos poner 2 graficas al mismo tiempo para compararlas*

*`Metodo 1 Paralelas`*

```python
import numpy as np


tiempo_2 = np.linspace(1, 10, 10)
velocidad_2 = np.array([0, 2, 4, 6, 1, 5, 7, 14, 3, 20])


# Tamaño de la figura
plt.figure(figsize=(10,6))

# Nacional Jugadores overall comparación de su tendencia
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primera posición ,Con esto logro que queden paralelos 3 parametro
plt.plot(tiempo,velocidad,color='red')
plt.title("Grafico 1")
plt.xlabel('Tiempo')
plt.ylabel("Velocidad")

# City Jugadores overall comparación de su tendencia
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segunda posición
plt.plot(tiempo_2,velocidad_2,color='orange')
plt.title("Grafico 2")
plt.xlabel('Tiempo')
plt.ylabel("Velocidad")

# Ajustar el diseño para evitar sobreposiciones
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

`Metodo 2 en la misma grafica`

```python
plt.figure(figsize=(10,6))

plt.plot(tiempo,velocidad,color='red',label='Grafica 1')

plt.plot(tiempo_2,velocidad_2,color='orange',label='Grafica 2')

plt.xlabel('Tiempo')

plt.ylabel("Velocidad")

plt.title('Graficas mezcladas')

plt.legend()

plt.show()

```
